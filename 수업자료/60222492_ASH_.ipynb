```python
#1.
a=(lambda x,y: x * y)(10, 20)
print(a)

#2.
def logged(func):
    def wrapper(*args, **kwargs):
        print('Calling add', func.__name__)
        return func(*args, **kwargs)   
    return wrapper
print("Calling add")
def add(x, y):
    return x + y

print(add(5,8))

#3.
import datetime 
print(str(datetime.datetime.now().isoformat()))

#4.
import datetime
threehundred = datetime.timedelta(days = 300)
how_long= datetime.datetime.now() - threehundred
print(how_long)

#5
numbers = [17,15,9,21,53,20,39,32,57,20,3,43,10]
max_value = None
max_idx = None

for idx, num in enumerate(numbers):
    if (max_value is None or num > max_value):
        max_value = num
        max_idx = idx

print('Maximum value:', max_value, "At index: ", max_idx)

#6
test_list = [8,7,3,-2,9]
def multiply(numbers):
    total = 1
    for x in numbers:
        total*= x
    return total
print(multiply(test_list))

#7
string = '1234abcdefg'
def string_reverse(str1):

    rstr1 = ''
    index = len(str1)
    while index > 0:
        rstr1 += str1[ index - 1 ]
        index = index - 1
    return rstr1
print(string_reverse('1234abcd'))


#8
class my_powcal:
  def mypow(self, x, n):
      if n==1 :
       return x
      if x == -1:
        if n%2 ==0:
           return -x
        else:
           return x
      if n==0:
         return 1
      if n<0:
         return 1/x,n
      val = self.mypow(x,n//2)
      if n%2 ==0:
         return val*val
      return x**n
print(my_powcal().pow(2, -3))
print(my_powcal().pow(3, 5))
print(my_powcal().pow(100, 0))

#9
class Rectangle():
 def __init__(self, h, w):
    self.h = h
    self.w = w
 def rectangle_area(self):
  return self.h*self.w
newRectangle = Rectangle(12, 14)
print(newRectangle.rectangle_area())

#10
from random import randrange
from datetime import timedelta

def random_date(start, end):
   
    delta = end - start
    int_delta = (delta.days * 24 * 60 * 60) + delta.seconds
    random_second = randrange(int_delta)
    return start + timedelta(seconds=random_second)


from datetime import datetime
d1 = datetime.strptime('5/25/2022', '%m/%d/%Y')
d2 = datetime.strptime('6/5/2022', '%m/%d/%Y')

print(random_date(d1, d2))

#11
fruits = ['Apple', 'Orange', 'Banana', 'Grape', 'Strawberry',
'Melon']
with open('fruit.txt') as myfile:
 for c in fruits:
  myfile.write(sorted(fruits))
content = open('fruit.txt')
print(content.read())

#12
def file_lengthy(fname):
 with open(fname) as f:
  for i, l in enumerate(f):
   pass
 return 
print("Number of lines in the file: ",file_lengthy("test.txt"))


#13

```
